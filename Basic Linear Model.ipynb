```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def estimate_coeff(x,y):
  n = np.size(x)
  m_x = np.mean(x)
  m_y = np.mean(y)
  Ss_xy = np.sum(y*x) - n*m_y*m_x
  Ss_xx = np.sum(x*x) - n*m_x*m_x
  b1 = Ss_xy/Ss_xx
  b0 = m_y - b1*m_x
  return (b0,b1)
```

```python
def plot_residuals(x, y, b):
    y_pred = b[0] + b[1]*x
    residuals = y - y_pred
    plt.scatter(x, residuals, color="red")
    plt.axhline(y=0, color='black', linestyle='--')
    plt.xlabel("House Size")
    plt.ylabel("Residuals")
    plt.title("Residual Plot")
    plt.show()
```

```python
def plot_regression_line(x,y,b):
  plt.scatter(x,y,color = "m",marker="o", s=30)
  y_pred = b[0] + b[1]*x
  plt.plot(x,y_pred,color = "g")
  plt.xlabel('House Size')
  plt.ylabel('Price')
  plt.show()
```

```python
def main():
    x = np.array([1500, 1700, 2100, 2400, 3000])
    y = np.array([300000, 320000, 340000, 360000, 400000])
    b0, b1 = estimate_coeff(x, y)
    print(f"The slope (b1 = {b1:.2f})")
    print(f"The intercept (b0 = {b0:.2f})")
    predicted_price = b0 + b1 * 2500
    print(f"Predicted price for 2500 sq ft house: ${predicted_price:,.2f}")
    plot_regression_line(x,y,(b0,b1))
    plot_residuals(x, y, (b0, b1))
if __name__ == "__main__":
    main()
```

Using Scikit Learn

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
x = np.array([1500, 1700, 2100, 2400, 3000])
y = np.array([300000, 320000, 340000, 360000, 400000])
x_reshaped = x.reshape(-1,1)
model.fit(x_reshaped,y)

print("\nResult from Scikit-learn")
print(f"Intercept (b0): {model.intercept_:.2f}")
print(f"Slope (b1) : {model.coef_[0]:.2f}")
predicted_price = model.predict([[2500]])
print(f"Predicted price for 2500 sq ft house: ${predicted_price[0]:,.2f}")
```
